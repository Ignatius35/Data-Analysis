### COLUMNS

##### it is better to put the columns in ["column1"] because alot of times columns have names so it makes it easy to add the names

```python
first_import["Column1"]
```

##### we can creat a new column as follows

```python
first_import["New_Column"] = 1
first_import["New_Column 2"] = list(range(10))
first_import
```

```python
first_import["New_Column"] = 1
first_import["New_Column 2"] = list(range(10))
first_import["Add to"] = first_import["New_Column"] + first_import["New_Column 2"]  # u can also use -, /,*, or %.
first_import 
```

```python
first_import = first_import.drop(columns="New_Column")  # drop usually used to take down a column
first_import
```

### ROWS
##### In the rows we use something called the "loc" and the "iloc" 
##### The loc methods allows us to use the names of columns to specify them

```python
first_import.loc[:5, ["Column1", "Column2", "Column3", "New_Column 2", "Add to"]] # Loc will include the last index you specify in this case 5
```

```python
first_import.iloc[:6, [0, 2]] #  iloc usually skips the first column and gives you the next column as specified in the second argument
```
